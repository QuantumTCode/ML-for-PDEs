## environment 

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import matplotlib
import matplotlib.pyplot
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import os
import itertools
import sys
```

```python
path = os.getcwd() 
parent_path = os.path.abspath(os.path.join(path, os.pardir))
sys.path.append(parent_path)
case_block = '/fdblock/'
case_filter = '/fdfilter/'
DF = []
testDF = []
testDF2 = []
scatterDF = []
```

```python
def Union(lst1, lst2): 
    final_list = list(set(lst1) | set(lst2)) 
    return final_list 

def fillup(fulllst):
    
    for ind in xrange(1,len(fulllst),1):
        if fulllst[ind] not in cost_epoch:
            cost_epoch.insert(ind,fulllst[ind])
            loss.insert(ind,loss[ind-1])
            oneMSE.insert(ind,oneMSE[ind-1])
            tenMSE.insert(ind,tenMSE[ind-1])
            sqMSE.insert(ind,sqMSE[ind-1])
    return cost_epoch,loss,oneMSE,tenMSE,sqMSE
```

## load results for FD-Blocks experiments

```python
algo = 'TR'
gap = 1

folder =  parent_path+'/Pretrained'+case_block 
resultsfiles = folder+'file.tar'

resultfilelist=torch.load(resultsfiles)['result']
bestfilelist=torch.load(resultsfiles)['best']

COST_EPOCH = {}
for f in resultfilelist:
    
    f = folder+f
    temp = torch.load(f)
    fb = temp['parm'][3]
    cost_iter = temp['Compute_Iter2']
    cost_epoch = list(np.array(cost_iter)+1.0)
    
    if fb not in COST_EPOCH:
        COST_EPOCH[fb] = cost_epoch
    else:
        COST_EPOCH[fb] = Union(COST_EPOCH[fb],cost_epoch)

    
for fb in COST_EPOCH:
    COST_EPOCH[fb].sort()

        
for f in resultfilelist:
    
    f = folder+f
    
    temp = torch.load(f)
    cost_iter = temp['Compute_Iter2']
    bs = temp['parm'][1]
    fr = temp['parm'][2]
    fb = temp['parm'][3]
    if fb < 10:
        fb = '#0%s'%temp['parm'][3]
    else:
        fb = '#%s'%temp['parm'][3]
    seed = temp['parm'][0]
    cost_epoch = list(np.array(cost_iter)+1.0)
    
    loss = temp['Loss']
    loss = [j if np.isnan(j)==False else np.inf for j in loss]
    loss = [j if j<np.inf else np.nan_to_num(j) for j in loss]
    
    oneMSE = temp['oneMSE']
    oneMSE = [j if np.isnan(j)==False else np.inf for j in oneMSE]
    oneMSE = [j if j<np.inf else np.nan_to_num(j) for j in oneMSE]
    
    tenMSE = temp['tenMSE']
    tenMSE = [j if np.isnan(j)==False else np.inf for j in tenMSE]
    tenMSE = [j if j<np.inf else np.nan_to_num(j) for j in tenMSE]
    
    sqMSE = temp['sqMSE']
    sqMSE = [j if np.isnan(j)==False else np.inf for j in sqMSE]
    sqMSE = [j if j<np.inf else np.nan_to_num(j) for j in sqMSE]
    
    cost_epoch = [cost_epoch[j] for j in xrange(0,len(cost_epoch),gap)]
    loss = [loss[j] for j in xrange(0,len(loss),gap)]
    oneMSE = [oneMSE[j] for j in xrange(0,len(oneMSE),gap)]
    tenMSE = [tenMSE[j] for j in xrange(0,len(tenMSE),gap)]
    sqMSE = [sqMSE[j] for j in xrange(0,len(sqMSE),gap)]
    

    scatterDF+=[[algo,seed,bs,fr,fb,cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i] \
        for cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i in \
        zip(cost_epoch,loss,oneMSE,tenMSE,sqMSE)]


    max_cost_epoch = cost_epoch[-1]
    
    min_loss_list = [j if np.isnan(j)==False else np.inf for j in loss]
    min_loss = np.min([j if j<np.inf else np.nan_to_num(j) for j in min_loss_list])

    min_sqMSE_list = [j if np.isnan(j)==False else np.inf for j in sqMSE]
    min_sqMSE = np.min([j if j<np.inf else np.nan_to_num(j) for j in min_sqMSE_list])
    min_oneMSE_list = [j if np.isnan(j)==False else np.inf for j in oneMSE]
    min_oneMSE = np.min([j if j<np.inf else np.nan_to_num(j) for j in min_oneMSE_list])
    min_tenMSE_list = [j if np.isnan(j)==False else np.inf for j in tenMSE]
    min_tenMSE = np.min([j if j<np.inf else np.nan_to_num(j) for j in min_tenMSE_list])

    fb = temp['parm'][3]
    testDF+=[[algo,seed,bs,fr,fb,max_cost_epoch,min_loss,min_oneMSE,min_tenMSE,min_sqMSE]]
    
    
    min_loss_epoch = cost_epoch[min_loss_list.index(min_loss)]
    min_oneMSE_epoch = cost_epoch[min_oneMSE_list.index(min_oneMSE)]
    min_tenMSE_epoch = cost_epoch[min_tenMSE_list.index(min_tenMSE)]
    min_sqMSE_epoch = cost_epoch[min_sqMSE_list.index(min_sqMSE)]
    
    testDF2+=[[algo,seed,bs,fr,fb,'loss',min_loss,min_loss_epoch]]
    testDF2+=[[algo,seed,bs,fr,fb,'one',min_oneMSE,min_oneMSE_epoch]]
    testDF2+=[[algo,seed,bs,fr,fb,'ten',min_tenMSE,min_tenMSE_epoch]]
    testDF2+=[[algo,seed,bs,fr,fb,'seq',min_sqMSE,min_sqMSE_epoch]]
    
    cost_epoch,loss,oneMSE,tenMSE,sqMSE = fillup(COST_EPOCH[fb])
    
    if fb < 10:
        fb = '#0%s'%temp['parm'][3]
    else:
        fb = '#%s'%temp['parm'][3]

    DF+=[[algo,seed,bs,fr,fb,cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i] \
        for cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i in \
        zip(cost_epoch,loss,oneMSE,tenMSE,sqMSE)]
```

## generate pandas dataframe

```python
df = pd.DataFrame(data=DF,\
                  columns=['algo','seed','bs','fr','FD-Block','Num of Computations','Loss',\
                           'one-step','ten-step','seq'])
scatterdf = pd.DataFrame(data=scatterDF,\
                  columns=['algo','seed','bs','fr','FD-Block','Num of Computations','Loss',\
                           'one-step','ten-step','seq'])
testdf = pd.DataFrame(data=testDF,\
                      columns=['algo','seed','bs','fr','FD-Block','cost','loss','one-step','ten-step','seq'])

testdf2 = pd.DataFrame(data=testDF2,\
                      columns=['algo','seed','bs','fr','FD-Block','','MSE','cost'])
```

### Figure 6

```python
from matplotlib.lines import Line2D

FB = [14,12,10,8,6,4,2]
sns.set()
sns.set(context="paper",font='serif')
sns.set_style("ticks")
fig, ax = plt.subplots(1, 2,figsize=(10,4),sharex=True,sharey=True)
ind = 0

palette = sns.color_palette()[:7]
markers = ('o', 'v', '^', '<', '>', '8', 's')
dashes = ("","","","","","","")
dashes = [(2,2)]*7

legend_elements = [Line2D([0],[0], color=palette[legend_ind], marker = markers[legend_ind], \
                          dashes=(2,2),lw=2,markersize=6)\
                   for legend_ind in xrange(7)]

lp=sns.lineplot(x='Num of Computations',y='Loss',dashes=dashes,palette=palette,\
             markers=markers,zorder=-1,markevery=10,markersize=5,legend='brief',hue='FD-Block',ci=95,style='FD-Block',\
                data=df,ax=ax[0])

ax[0].set_yscale('log')
ax[0].set_xscale('log')
ax[0].set_title('Training',fontsize=20,pad=10)
ax[0].set_yscale('log')
ax[0].set_xscale('log')
ax[0].set_xlim([100.0,20000.0])
ax[0].set_ylabel('MSE',fontsize=20,labelpad=10)
ax[0].set_xlabel('')
handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend(handles=handles[:1]+legend_elements, \
             labels=labels, loc='upper left',fontsize=12, fancybox=True, framealpha=0.5)

lp=sns.lineplot(x='Num of Computations',y='seq',dashes=dashes,palette=palette,\
             markers=markers,zorder=-1,markevery=10,markersize=5,legend=False,hue='FD-Block',ci=95,style='FD-Block',\
                data=df,ax=ax[1])
ax[1].set_yscale('log')
ax[1].set_xscale('log')
ax[1].set_title('1000-Step Prediction',fontsize=20,pad=10)
ax[1].set_yscale('log')
ax[1].set_xscale('log')
ax[1].set_ylim([1e-13,1e+9])
ax[1].set_xlabel('')
ax[1].set_ylabel('')

for i in xrange(2):
    ax[i].tick_params(labelsize=20,length=1.0)
fig.text(0.26, -0.025, '# gradient and Hessian-vector product', va='center',fontsize=20)
fig.tight_layout()
fig.savefig('block_train_pred_traject_new.pdf',bbox_inches = "tight")
plt.show()
```
