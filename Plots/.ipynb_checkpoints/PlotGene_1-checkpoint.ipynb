# This notebook generates the following figures for our main paper:
### Figure 2. Evolution of training errors
### Figure 5. Testing errors (minimum vs final)

## environment 

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import matplotlib
import matplotlib.pyplot
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
import os
import itertools
import sys
```

```python
path = os.getcwd() 
parent_path = os.path.abspath(os.path.join(path, os.pardir))
sys.path.append(parent_path)
DO = [0,1,2,3] # stable, unstable, noisy-medium, force
CASE = ['/stable/','/unstable/','/midnoise/','/force/']
DF = []
testDF = []
testDF2 = []
scatterDF = []
```

## load results TR 

```python
Algo = 'TR/'
for data_option in DO:
    
    fr = 16 # number of FD-Filters
    bs = 64 # mini-batch size
    
    if data_option == 1:
        fb = 10 # number of FD-Blocks for unstable case
    else:
        fb = 1 # number of FD-Blocks for other cases
    
    case = CASE[data_option]
    folder = parent_path+'/Pretrained'+case+Algo
    
        
    print data_option,fb,fr,folder
    
    resultsfiles = folder+'file.tar'
    
    resultfilelist=torch.load(resultsfiles)['result']
    bestfilelist=torch.load(resultsfiles)['best']

    for f in resultfilelist:
        
        f = folder+f
        algo = 'TR'
        temp = torch.load(f)
        cost_iter = temp['Compute_Iter2']
        bs = temp['parm'][1]
        fr = temp['parm'][2]
        fb = temp['parm'][3]
        seed = temp['parm'][0]
        cost_epoch = list(np.array(cost_iter)+1.0)

        loss = temp['Loss']
        loss = [j if np.isnan(j)==False else np.inf for j in loss]
        loss = [j if j<np.inf else np.nan_to_num(j) for j in loss]

        oneMSE = temp['oneMSE']
        oneMSE = [j if np.isnan(j)==False else np.inf for j in oneMSE]
        oneMSE = [j if j<np.inf else np.nan_to_num(j) for j in oneMSE]

        tenMSE = temp['tenMSE']
        tenMSE = [j if np.isnan(j)==False else np.inf for j in tenMSE]
        tenMSE = [j if j<np.inf else np.nan_to_num(j) for j in tenMSE]

        sqMSE = temp['sqMSE']
        sqMSE = [j if np.isnan(j)==False else np.inf for j in sqMSE]
        sqMSE = [j if j<np.inf else np.nan_to_num(j) for j in sqMSE]

        DF+=[[data_option,algo,seed,bs,fr,fb,cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i] \
            for cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i in \
            zip(cost_epoch,loss,oneMSE,tenMSE,sqMSE)]

        scatterDF+=[[data_option,algo,seed,bs,fr,fb,cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i] \
            for cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i in \
            zip(cost_epoch,loss,oneMSE,tenMSE,sqMSE)]


        max_cost_epoch = cost_epoch[-1]
        min_sqMSE = np.min(sqMSE)
        min_oneMSE = np.min(oneMSE)
        min_tenMSE = np.min(tenMSE)
        
        end_cost_epoch = cost_epoch[-1]
        end_sqMSE = sqMSE[-1]
        end_oneMSE = oneMSE[-1]
        end_tenMSE = tenMSE[-1]

        testDF+=[[data_option,algo,seed,bs,fr,fb,max_cost_epoch,min_oneMSE,min_tenMSE,min_sqMSE]]
        testDF2+=[[data_option,algo,seed,bs,fr,fb,max_cost_epoch,end_oneMSE,end_tenMSE,end_sqMSE]]
```

## load results ADAM

```python
Algo = 'Adam/'
for data_option in DO:
    
    fr = 16 # number of FD-Filters
    bs = 64 # mini-batch size
    
    if data_option == 1:
        fb = 10 # number of FD-Blocks for unstable case
    else:
        fb = 1 # number of FD-Blocks for other cases
    
    case = CASE[data_option]
    folder = parent_path+'/Pretrained'+case+Algo
    
        
    print data_option,fb,fr,folder
    
    resultsfiles = folder+'file.tar'
    
    resultfilelist=torch.load(resultsfiles)['result']
    bestfilelist=torch.load(resultsfiles)['best']

    for f in resultfilelist:
        
        f = folder+f
            
        best_f = f[:-11]+'BEST.tar'
        if os.path.exists(best_f)==False:

            if f[:-12][-6:]=='0.0001':
                lr = 0.0001
            else:
                lr = 0.001
        else:
            lr = torch.load(best_f)['parm'][4]
        algo='A-%.e'%lr
        
        temp = torch.load(f)
        cost_iter = range(0,12001,30)
        bs = temp['parm'][1]
        fr = temp['parm'][2]
        fb = temp['parm'][3]
        seed = temp['parm'][0]
        cost_epoch = list(np.array(cost_iter)+1.0)

        loss = temp['Loss']
        loss = [j if np.isnan(j)==False else np.inf for j in loss]
        loss = [j if j<np.inf else np.nan_to_num(j) for j in loss]

        oneMSE = temp['oneMSE']
        oneMSE = [j if np.isnan(j)==False else np.inf for j in oneMSE]
        oneMSE = [j if j<np.inf else np.nan_to_num(j) for j in oneMSE]

        tenMSE = temp['tenMSE']
        tenMSE = [j if np.isnan(j)==False else np.inf for j in tenMSE]
        tenMSE = [j if j<np.inf else np.nan_to_num(j) for j in tenMSE]

        sqMSE = temp['sqMSE']
        sqMSE = [j if np.isnan(j)==False else np.inf for j in sqMSE]
        sqMSE = [j if j<np.inf else np.nan_to_num(j) for j in sqMSE]

        DF+=[[data_option,algo,seed,bs,fr,fb,cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i] \
            for cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i in \
            zip(cost_epoch,loss,oneMSE,tenMSE,sqMSE)]


        max_cost_epoch = cost_epoch[-1]
        min_sqMSE = np.min(sqMSE)
        min_oneMSE = np.min(oneMSE)
        min_tenMSE = np.min(tenMSE)

        end_cost_epoch = cost_epoch[-1]
        end_sqMSE = sqMSE[-1]
        end_oneMSE = oneMSE[-1]
        end_tenMSE = tenMSE[-1]

        testDF+=[[data_option,algo,seed,bs,fr,fb,max_cost_epoch,min_oneMSE,min_tenMSE,min_sqMSE]]
        testDF2+=[[data_option,algo,seed,bs,fr,fb,max_cost_epoch,end_oneMSE,end_tenMSE,end_sqMSE]]


        total = len(loss)
        inter = (total-1)/100

        cost_epoch = [cost_epoch[j] for j in xrange(0,total,inter)]
        loss = [loss[j] for j in xrange(0,total,inter)]
        oneMSE = [oneMSE[j] for j in xrange(0,total,inter)]
        tenMSE = [tenMSE[j] for j in xrange(0,total,inter)]
        sqMSE = [sqMSE[j] for j in xrange(0,total,inter)]


        scatterDF+=[[data_option,algo,seed,bs,fr,fb,cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i] \
            for cost_epoch_i,loss_i,oneMSE_i,tenMSE_i,sqMSE_i in \
            zip(cost_epoch,loss,oneMSE,tenMSE,sqMSE)]
```

# generate pandas dataframe 

```python
df = pd.DataFrame(data=DF,\
                  columns=['case','algo','seed','bs','fr','fb','Num of Computations','Loss',\
                           'one-step','ten-step','seq'])
scatterdf = pd.DataFrame(data=scatterDF,\
                  columns=['case','algo','seed','bs','fr','fb','Num of Computations','Loss',\
                           'one-step','ten-step','seq'])
testdf = pd.DataFrame(data=testDF,\
                      columns=['case','algo','seed','bs','fr','fb','cost','one-step','ten-step','seq'])

testdf2 = pd.DataFrame(data=testDF2,\
                      columns=['case','algo','seed','bs','fr','fb','cost','one-step','ten-step','seq'])

FR = [16,16,16,16]
FB = [1,10,1,1]
Case = ['Stable', 'Unstable', 'Noisy: Medium', 'Forcing']
```

### Figure 2 

```python
from matplotlib.lines import Line2D as L

sns.set()
sns.set(context="paper",font='serif')
sns.set_style("ticks")
fig, ax = plt.subplots(2, 2,figsize=(4,4),sharex=True,sharey=True)
ind = 0

case_ind = 0

for i,j in itertools.product(range(2),range(2)):
    if i==0 and j==0:
        LEGEND='full'
    else:
        LEGEND=False
    subdf = df[df['case']==case_ind]
    lp = sns.lineplot(x='Num of Computations',y='Loss',dashes=True,markers=True,zorder=-1,\
                      markevery=10,legend=LEGEND,hue='algo',\
                 style='algo',ci=95,\
                data=subdf,ax=ax[i][j])
    ax[i][j].set_title(Case[case_ind])
    ax[i][j].set_yscale('log')
    ax[i][j].set_xscale('log')
    ax[i][j].set_xlim([1.0,12001.0])
    ax[i][j].set_ylim([1e-14,1e+5])
    ax[i][j].set_ylabel('')
    ax[i][j].set_xlabel('')
    if i==0 and j==0:
        handles, labels = ax[i][j].get_legend_handles_labels()
        ax[i][j].legend(handles[1:], labels[1:], loc='upper left')
#         ax[i][j].set_ylabel('MSE')
#     elif j==0:
#         ax[i][j].set_ylabel('MSE')
        
    case_ind+=1
fig.text(0.0, 0.5, 'MSE', rotation=90,va='center')
fig.text(0.2, 0.0, '# gradient and Hessian-vector product', va='center')
fig.tight_layout()
fig.savefig('loss_by_case.pdf',bbox_inches = "tight")
plt.show()
```

### Figure 5

```python
sns.set()
sns.set(context="paper",font='serif')
sns.set_style("whitegrid")
fig, ax = plt.subplots(2, 2,figsize=(4,4),sharex=True,sharey=True)

case_ind = 0
for i,j in itertools.product(range(2),range(2)):
    subdf = testdf[testdf['case']==case_ind]
    ax[i][j].set_title(Case[case_ind])
    ax[i][j].set_yscale('log')
    box=sns.boxplot(x="algo",y="seq",whis='range',width=0.4,data=subdf,ax=ax[i][j])
    swarm=sns.swarmplot(x="algo", y="seq",data=subdf,size=5, zorder=-1,linewidth=1,ax=ax[i][j])
    ax[i][j].set_xlabel('')
#     if j==0:
#         ax[i][j].set_ylabel('Min MSE')
#     else:
    ax[i][j].set_ylabel('')
    ax[i][j].set_ylim([1e-11,1e+18])
    case_ind+=1
fig.text(0.0, 0.5, 'Min MSE', rotation=90,va='center')
fig.tight_layout()
fig.savefig('1000_step_test_min.pdf',bbox_inches = "tight")
plt.show()

fig, ax = plt.subplots(2, 2,figsize=(4,4),sharex=True,sharey=True)

case_ind = 0
for i,j in itertools.product(range(2),range(2)):
    subdf = testdf2[testdf2['case']==case_ind]
    ax[i][j].set_title(Case[case_ind])
    ax[i][j].set_yscale('log')
    box=sns.boxplot(x="algo",y="seq",whis='range',width=0.4,data=subdf,ax=ax[i][j])
    swarm=sns.swarmplot(x="algo", y="seq",data=subdf,size=5, zorder=-1,linewidth=1,ax=ax[i][j])
    ax[i][j].set_xlabel('')
#     if i==0:
#         ax[i][j].set_ylabel('End MSE')
#     else:
    ax[i][j].set_ylabel('')
    ax[i][j].set_ylim([1e-11,1e+18])
    case_ind+=1
fig.text(0.0, 0.5, 'Final MSE', rotation=90,va='center')
fig.tight_layout()
fig.savefig('1000_step_test_final.pdf',bbox_inches = "tight")
plt.show()
```
